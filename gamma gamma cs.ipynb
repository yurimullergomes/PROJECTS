```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
#the data below is the QED data
df_crosssection = pd.read_csv("secaodechoque1.txt", delimiter='\t')

```

```python
df_cs = df_crosssection.transpose()
```

```python
df_cs[0]
```

```python
#Y= df_cs[3]
X = [i for i in range(100)]
#plt.plot(X,Y)
#plt.show()

for j in range(99):
    Y = df_cs[j]
    plt.plot(X, Y)
    plt.title("$\gamma \gamma$ differential Cross section")
    plt.xlabel("scattering Angle $\Theta$ ")
    plt.ylabel("$ \partial\sigma/\partial \Theta $")
#    plt.legend()
#legend = plt.legend(loc='upper right', shadow=True, fontsize='small')

plt.show()
```

```python
integral = []
for i in range(98):
    integral.append(df_cs[i].sum())
print(integral)
```

```python
X = [i/40 for i in range(98)]
plt.plot(X,integral)
plt.title("$\gamma \gamma$ Total Cross section")
plt.xlabel("Energy (MeV)")
plt.ylabel("$\sigma$")

plt.show()

```
